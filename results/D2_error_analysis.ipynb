```python
import pandas as pd
import json

# show full text in dataframe display
pd.set_option('display.max_colwidth', None) 
```

```python
# read in predictions
with open("../outputs/D2/d2.out", 'r') as f:
    i = 0
    preds = []
    golds = []
    for line in f.readlines():
        split_line = line.split(",")
        preds.append(split_line[0].split(": ")[1])
        # golds.append(split_line[1].split(": ")[1])

# data = {"predictions" : preds, "gold_labels" : golds}
data = {"predictions" : preds}
```

```python
# make predictions into df
pred_df = pd.DataFrame.from_dict(data)
pred_df.head()
```

```python
# open dev data
with open("../data/sarc/dev-comments-balanced.json", 'r') as f:
    test_data_raw = json.load(f)
```

```python
# make dev dataframe
dev_df = pd.DataFrame(test_data_raw)
dev_df["predictions"] = pred_df["predictions"].astype(int)
dev_df["label"] = dev_df["label"].astype(int)
dev_df.head()
```

```python
# create false neg/false pos
false_neg = dev_df[(dev_df["label"] == 0) & (dev_df["predictions"] == 1)]
false_pos = dev_df[(dev_df["label"] == 1) & (dev_df["predictions"] == 0)]
```
